```python
from google.colab import drive

drive.mount('/content/gdrive')

import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/디지털 영상처리 강의 관련/Images/scanimg.jpg')

#print(img.shape)
#cv2_imshow(img)
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#plt.imshow(RGB_img)

gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
plt.imshow(gimg, cmap='gray')
```

```python
kernel = np.ones((3, 3), np.uint8) # 3x3의 구조요소
erodeImg = cv2.erode(gimg, kernel, iterations = 2)
cv2_imshow(erodeImg)
```

```python
kernel = np.ones((3, 3), np.uint8) # 3x3의 구조요소
dilateImg = cv2.dilate(gimg, kernel, iterations = 1)
cv2_imshow(dilateImg)
```

```python
kernel = np.ones((4, 1), np.uint8) # 구조 요소를 조정
dilateImg = cv2.dilate(gimg, kernel, iterations = 1)
cv2_imshow(dilateImg)
```

```python
kernel = np.ones((4, 1), np.uint8) # 복원
erodeImg = cv2.erode(dilateImg, kernel, iterations = 1)
cv2_imshow(erodeImg)
```
